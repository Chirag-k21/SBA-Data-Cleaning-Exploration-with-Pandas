```python
#Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset

df1 = pd.read_excel("Table of Size Standards_Effective May 2 2022_Final.xlsx", sheet_name = 'table_of_size_standards-all')
```

```python
df1.head()
```

```python
df1.info()
```

```python
#looking for null values
df1.isnull().sum()
```

```python
#Selecting NAICS Industry Description column to extract sector and sector codes
df = df1[df1['NAICS Codes'].isnull()]
df
```

# DATA CLEANING

```python
#Extracting sector codes and sector description
df['Sector'] = df['NAICS Industry Description'].str[7:9]
df['Sector_Description'] = df['NAICS Industry Description'].str.slice(start = 12, stop = 538)
df
```

```python
df = df[['NAICS Industry Description', 'Sector', 'Sector_Description']]
df
```

```python
#Removing unrequired rows containing misleading info
df = df[df['Sector'].str.isnumeric()]
df
```

```python
#Replacing some values in sector description column

df.loc[ df['Sector_Description'] == '33 – Manufacturing', 'Sector_Description'] = 'Manufacturing'
df
```

```python
df['Sector_Description'] = np.where(df['Sector_Description'] == '45 – Retail Trade', 'Retail Trade', df['Sector_Description'])
df['Sector_Description'] = np.where(df['Sector_Description'] == '49 – Transportation and Warehousing', 'Transportation and Warehousing', df['Sector_Description'])
df
```

```python
#Replacing some values in the dataset to compensate combined sectors
a1 = {'NAICS Industry Description': 'Sector 31 – 33 – Manufacturing', 'Sector': 32, 'Sector_Description': 'Manufacturing'}
a2= {'NAICS Industry Description': 'Sector 31 – 33 – Manufacturing', 'Sector': 33, 'Sector_Description': 'Manufacturing'}
a3= {'NAICS Industry Description': 'Sector 44 - 45 – Retail Trade', 'Sector': 45, 'Sector_Description': 'Retail Trade'}
a4= {'NAICS Industry Description': 'Sector 48 - 49 – Transportation and Warehousing', 'Sector': 49, 'Sector_Description': 'Transportation and Warehousing'}

df = df.append([a1,a2,a3,a4], ignore_index = True) 
```

```python
df
```

```python
df['Sector'] = df['Sector'].astype('int64')
df = df.sort_values(['Sector'])
df
```

```python
df.to_csv('Clean_SBA_data.csv', encoding = 'utf-8', index = False)
```
